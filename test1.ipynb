```python
def func(x,dic):
    factor = ''
    if bin(x & 0b1) == bin(2**0):
        if len(factor)==0:
            factor += dic[1]
        elif factor in 'Unknown':
            factor = dic[1]
        elif dic[1] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[1]
    if bin(x & 0b10) == bin(2**1):
        if len(factor)==0:
            factor += dic[2]
        elif factor in 'Unknown':
            factor = dic[2]
        elif dic[2] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[2]
    if bin(x & 0b100) == bin(2**2):
        if len(factor)==0:
            factor += dic[3]
        elif factor in 'Unknown':
            factor = dic[3]
        elif dic[3] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[3]
    if bin(x & 0b1000) == bin(2**3):
        if len(factor)==0:
            factor += dic[4]
        elif factor in 'Unknown':
            factor = dic[4]
        elif dic[4] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[4]
    if bin(x & 0b10000) == bin(2**4):
        if len(factor)==0:
            factor += dic[5]
        elif factor in 'Unknown':
            factor = dic[5]
        elif dic[5] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[5]
    if bin(x & 0b100000) == bin(2**5):
        if len(factor)==0:
            factor += dic[6]
        elif factor in 'Unknown':
            factor = dic[6]
        elif dic[6] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[6]
    if bin(x & 0b1000000) == bin(2**6):
        if len(factor)==0:
            factor += dic[7]
        elif factor in 'Unknown':
            factor = dic[7]
        elif dic[7] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[7]
    if bin(x & 0b10000000) == bin(2**7):
        if len(factor)==0:
            factor += dic[8]
        elif factor in 'Unknown':
            factor = dic[8]
        elif dic[8] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[8]
    if bin(x & 0b100000000) == bin(2**8):
        if len(factor)==0:
            factor += dic[9]
        elif factor in 'Unknown':
            factor = dic[9]
        elif dic[9] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[9]
    if bin(x & 0b1000000000) == bin(2**9):
        if len(factor)==0:
            factor += dic[10]
        elif factor in 'Unknown':
            factor = dic[10]
        elif dic[10] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[10]
    if bin(x & 0b10000000000) == bin(2**10):
        if len(factor)==0:
            factor += dic[11]
        elif factor in 'Unknown':
            factor = dic[11]
        elif dic[11] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[11]
    if bin(x & 0b100000000000) == bin(2**11):
        if len(factor)==0:
            factor += dic[12]
        elif factor in 'Unknown':
            factor = dic[12]
        elif dic[12] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[12]
    if bin(x & 0b1000000000000) == bin(2**12):
        if len(factor)==0:
            factor += dic[13]
        elif factor in 'Unknown':
            factor = dic[13]
        elif dic[13] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[13]
    if bin(x & 0b10000000000000) == bin(2**13):
        if len(factor)==0:
            factor += dic[14]
        elif factor in 'Unknown':
            factor = dic[14]
        elif dic[14] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[14]
    if bin(x & 0b100000000000000) == bin(2**14):
        if len(factor)==0:
            factor += dic[15]
        elif factor in 'Unknown':
            factor = dic[15]
        elif dic[15] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[15]
    if bin(x & 0b1000000000000000) == bin(2**15):
        if len(factor)==0:
            factor += dic[16]
        elif factor in 'Unknown':
            factor = dic[16]
        elif dic[16] in 'Unknown':
            factor = factor
        else:
            factor += '+'+dic[16]
    return factor
    

df['CPR_Factor'] = df.error_code.apply(func, dic=factor_dict)
df
```
