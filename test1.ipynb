```python
import pandas as pd
import difflib

# CSVファイルを読み込む
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')

# 結果を保存するためのリスト
results = []

# モデル2のデータを繰り返し処理
for index, row in file2.iterrows():
    model2 = row['モデル２']
    digits_in_model2 = ''.join([char for char in model2 if char.isdigit()])
    
    # 数字が一致するモデル1を検索
    matched_rows = file1[file1['モデル１'].str.contains(digits_in_model2)]
    
    if not matched_rows.empty:
        # 複数の一致がある場合、全てのデータ1を連結
        data1_values = ','.join(matched_rows['データ１'].astype(str))
    else:
        # 部分一致を検索
        model1_candidates = file1['モデル１'].tolist()
        close_matches = difflib.get_close_matches(model2, model1_candidates, n=1, cutoff=0.5)
        if close_matches:
            # 最も近いモデル1を使用
            closest_model1 = close_matches[0]
            data1_values = file1[file1['モデル１'] == closest_model1]['データ１'].values[0]
        else:
            data1_values = ''
    
    results.append([model2, data1_values])

# 新しいCSVファイルに保存
output_df = pd.DataFrame(results, columns=['モデル２', 'データ１'])
output_df.to_csv('output.csv', index=False)
```

```python
import pandas as pd
import Levenshtein
import re

# 実際のデータに置き換え
data1 = {
    'ブランド 1': ['リコー', 'リコー', 'リコー', 'リコー', 'キャノン'],
    'モデル 1': ['Alicia 1013', 'Alicia 1013F', 'Alicia MP 9002', 'Alicia MP 9003', 'ImageRunner02']
}

data2 = {
    'ブランド 2': ['リコー', 'リコー', 'リコー', 'リコー', 'キャノン'],
    'モデル 2': ['AF1013', 'AF1013F', 'AFM9002', 'AFM9004', 'R02']
}

# データフレームに変換
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# 数字を抽出するための関数
def extract_numbers(model):
    return re.findall(r'\d+', model)

# 数字が完全に一致するか、または最後の一桁以外が一致するかを確認する関数
def numbers_match(numbers1, numbers2):
    if numbers1 == numbers2:
        return True
    if len(numbers1) == len(numbers2) and all(n1[:-1] == n2[:-1] for n1, n2 in zip(numbers1, numbers2)):
        return True
    return False

# 結果を格納するリスト
results = []

# マッチングプロセス
for idx2, row2 in df2.iterrows():
    brand2 = row2['ブランド 2']
    model2 = row2['モデル 2']
    model2_numbers = extract_numbers(model2)
    
    best_match = None
    best_score = float('inf')
    
    for idx1, row1 in df1.iterrows():
        if row1['ブランド 1'] == brand2:
            model1 = row1['モデル 1']
            model1_numbers = extract_numbers(model1)
            
            # 数字部分が一致するか、または最後の一桁以外が一致するかをチェック
            if numbers_match(model2_numbers, model1_numbers):
                score = Levenshtein.distance(model2, model1)
                if score < best_score:
                    best_score = score
                    best_match = model1
    
    results.append({
        'ブランド 2': brand2,
        'モデル 2': model2,
        'モデル 1': best_match,
        'Levenshtein 距離': best_score
    })

# 結果をデータフレームに変換
df_results = pd.DataFrame(results)

# 結果を表示
print(df_results)
```

```python
import pandas as pd
import Levenshtein as lev
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

def normalize_string(s):
    return ''.join(char.lower() for char in s if char.isalnum())

def calculate_similarity(s1, s2):
    lev_score = lev.ratio(s1, s2)
    vec = CountVectorizer().fit_transform([s1, s2])
    vectors = vec.toarray()
    cos_score = cosine_similarity(vectors)[0, 1]
    return (lev_score + cos_score) / 2

# CSVファイルの読み込み
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')

def find_best_matches(row, candidates, num_matches=3):
    model2_normalized = normalize_string(row.Model2)  # インデックスではなく属性としてアクセス
    scores = []

    for candidate in candidates.itertuples():
        model1_normalized = normalize_string(candidate.Model1)
        score = calculate_similarity(model2_normalized, model1_normalized)
        scores.append((candidate.Model1, score))
    
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:num_matches]

results = []

# Brand2とBrand1が一致する中で類似度を計算
for row in file2.itertuples():
    brand_matches = file1[file1['Brand1'] == row.Brand2]
    matches = find_best_matches(row, brand_matches)

    if matches:
        match_results = {
            'Model2': row.Model2,
            'Brand2': row.Brand2,
            'Match1': matches[0][0],
            'Score1': matches[0][1]
        }
        
        if len(matches) > 1:
            match_results.update({
                'Match2': matches[1][0],
                'Score2': matches[1][1]
            })
        if len(matches) > 2:
            match_results.update({
                'Match3': matches[2][0],
                'Score3': matches[2][1]
            })

        results.append(match_results)

output_df = pd.DataFrame(results)
output_df.to_csv('output.csv', index=False)
```

```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# データの読み込み
file1_df = pd.read_csv('file1.csv')
file2_df = pd.read_csv('file2.csv')

# Word2Vecモデルのトレーニング
all_models = pd.concat([file1_df['Model1'], file2_df['Model2']]).tolist()
all_models = [model.split() for model in all_models]
model = Word2Vec(sentences=all_models, vector_size=100, window=5, min_count=1, workers=4)

def get_vector(word):
    return model.wv[word] if word in model.wv else None

def get_similarity(model1, model2):
    vec1 = get_vector(model1)
    vec2 = get_vector(model2)
    if vec1 is not None and vec2 is not None:
        return cosine_similarity([vec1], [vec2])[0][0]
    return 0

results = []

for _, row in file2_df.iterrows():
    matches = []
    for _, file1_row in file1_df.iterrows():
        if row['Brand2'] == file1_row['Brand1']:
            similarity = get_similarity(row['Model2'], file1_row['Model1'])
            matches.append((file1_row['Model1'], similarity))
    matches.sort(key=lambda x: x[1], reverse=True)
    result = {
        'Model2': row['Model2'],
        'Brand2': row['Brand2']
    }
    for i, match in enumerate(matches[:3]):
        result[f'Match{i+1}'] = match[0]
        result[f'Score{i+1}'] = match[1]
    results.append(result)

output_df = pd.DataFrame(results)
output_df.to_csv('output.csv', index=False)

```
